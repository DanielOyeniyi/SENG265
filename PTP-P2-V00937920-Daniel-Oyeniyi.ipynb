### 19 Python list slicing and comprehension 

List slicing and comprehensions are easy way to copy lists in python. 

#### Slicing
Allows you to copy a section of a list, for example:

```python
x: list = [1, 2, 3, 4, 5]
y: list = x[2:4]
print(y)
```

#### Comprehension
Allows you to copy only the desired items in a list, for example:

```python
x: list = [2,8,3,5,10,1]
y: list = [num for num in x if num % 2 == 0]
print(y)
```

### 21 Deep Copy vs Shallow Copy

Shallow copy will copy the reference to the object whereas deep copy will copy each element of the object. So if you only perform a shallow copy then modifications to the reference object will still change both copies. You won't run into this problem with deep copy but it takes longer.

#### Shallow Copy

```python
import copy 
original: list = [1, 2, 3, 4, 5, 6, 7]
shallow: list = original # creates a shallow copy
print(original)
print(shallow)

original[2] = 1000
original.append(44)
print(shallow)
print(original)
```
